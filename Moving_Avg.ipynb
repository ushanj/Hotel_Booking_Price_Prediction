```python
# importing time-series data
df = pd.read_csv('Travel_scrape.csv', index_col='Search Time',
                       parse_dates=True)
 
# Printing dataFrame
df.head()
```

```python
df = df.drop(['Unnamed: 6'], axis=1)
df.head()
```

```python
df = df.astype(str)
df['Rate'] = df['Rate'].apply(lambda x: x[3:])
```

```python
# def filter_Hotel()
hotel_names = df['Hotel'].unique()
```

```python
hotel_names
```

```python

# selecting rows based on condition 
hotel_df = df.loc[df['Hotel'] == hotel_names[1]]
hotel_df.head()
```

```python
df.tail()
```

```python
df1 = df['Rate'].to_frame()

# calculating simple moving average
# using .rolling(window).mean() ,
# with window size = 30
df1['SMA30'] = df['Rate'].rolling(30).mean()

# removing all the NULL values using
# dropna() method
df1.dropna(inplace=True)

# printing Dataframe
df1
```

```python
df1 = df1.reset_index()
df1 = df1.rename(columns={"index":"Booking_ID"})
df1['Booking_ID'] = df1.index + 1
```

```python
# df1.reset_index(inplace = True, drop = True)
df1
```

```python
# plotting Close price and simple
# moving average of 30 days using .plot() method
df1['SMA30'].plot(label='df1',figsize=(16, 5))
```

```python
def plot_hotel_wise_MA(x):
    hotel_names = df['Hotel'].unique()
    hotel_df = df.loc[df['Hotel'] == hotel_names[x]]
    
    hotel_df = hotel_df['Rate'].to_frame()
    hotel_df['SMA30'] = hotel_df['Rate'].rolling(30).mean()
    hotel_df.dropna(inplace=True)
    
    hotel_df = hotel_df.reset_index()
    hotel_df = hotel_df.rename(columns={"index":"Booking_ID"})
    hotel_df['Booking_ID'] = hotel_df.index + 1
    
    return hotel_df['SMA30'].plot(label='hotel_df',figsize=(16, 5))

    
```

```python
plot_hotel_wise_MA(1)
```
